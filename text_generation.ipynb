```python
import fitz as f

def read_pdf (file_path):
  text = ''
  doc = f.open(file_path)
  for page in doc :
    text+= page.get_text()
  doc.close()
  return text
```

```python
main_text = read_pdf('news.pdf')
```

```python
def chunk_by_char (text, chunk_size = 300 ,overlap = 50 ):
  words = text.split()
  global chunks
  chunks = []

  for i in range(0 ,len(words) , chunk_size - overlap):
    chunk = words[i:i+chunk_size]
    chunks.append(' '.join(chunk))
  return chunks
```

```python
main_version = chunk_by_char(main_text)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
embedding = model.encode(main_version)
```

```python
import faiss
import numpy as np

embedding_dim = embedding[0].shape[0]
index = faiss.IndexFlatL2(embedding_dim)
embedding_np = np.array(embedding).astype('float32')
index.add(embedding_np)

query = 'what is the artifical inteligens ? '
embedding_query = model.encode([query]).astype('float32')
```

```python
k = 3
distace , indices = index.search(embedding_query,k)

print(f'the simmilar embedding is {indices}')
print(f'Distance {distace}')
```

```python
from transformers import pipeline

context = "\n".join([chunks[i] for i in indices[0]])

prompt = f"""
answer the qeustion based on the following context:
the context is {context}
and the question id {query}
"""
qa_pipeline = pipeline('text-generation',model="gpt2")

response = qa_pipeline(prompt , max_new_tokens=100 , do_sample=True)[0]['generated_text']
print(f'Answer :{response}')
```
